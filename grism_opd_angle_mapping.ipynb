```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import numpy.ma as ma
from scipy import misc, signal
from astropy.io import fits
```

```python
# the function of the angle at the top of the grism

def opd_top_grism(lambda_s, lambda_l, omega, del_y, D_s):
    '''
    INPUTS:
    lambda_s: the short wavelength limit (best to leave as um)
    lambda_l: the long wavelength limit
    omega: the angle with the vertical (in deg.), measured from the top (short-wavelength end) of the grism
    del_y: the span of the grism PSF, in pixels
    D_s: the distance in pixels spanned by one of the lambda_s
    
    OUTPUTS:
    opd: the OPD (in same units as the lambdas)
    '''
    
    opd = np.divide(np.power(lambda_s,2)*del_y,D_s*(lambda_l-lambda_s))*np.tan(omega*np.pi/180.)
    
    return opd
```

```python
# the function of the angle at the bottom of the grism

def opd_bottom_grism(lambda_s, lambda_l, omega, del_y, D_l):
    '''
    INPUTS:
    lambda_s: the short wavelength limit (best to leave as um)
    lambda_l: the long wavelength limit
    omega: the angle with the vertical (in deg.), measured from the top (short-wavelength end) of the grism
    del_y: the span of the grism PSF, in pixels
    D_l: the distance in pixels spanned by one of the lambda_l
    
    OUTPUTS:
    opd: the OPD (in same units as the lambdas)
    '''
    
    opd = np.divide(np.power(lambda_l,2)*del_y,D_l*(lambda_l-lambda_s))*np.tan(omega*np.pi/180.)
    
    return opd
```

```python
sample_angles = np.arange(-40,40,0.1)
sample_opd_top = opd_top_grism(lambda_s=2.8,lambda_l=4.2,omega=sample_angles,del_y=200,D_s=1)
sample_opd_bottom = opd_bottom_grism(lambda_s=2.8,lambda_l=4.2,omega=sample_angles,del_y=200,D_l=4.2/2.8)
```

```python
plt.plot(sample_angles,sample_opd_top)
plt.plot(sample_angles,sample_opd_bottom)
plt.xlabel("Fringe angle (deg)")
plt.ylabel("$\Delta$SPC position (motor counts)")
plt.show()
```

```python
class fft_img:
    # take FFT of a 2D image

    def __init__(self, image):
        self.image = image

    def fft(self, padding=int(0), pad_mode='constant', mask_thresh=1e-10, mask=True):

        padI = np.pad(self.image, padding, pad_mode)
        # arguments: image, pad size, pad mode, threshold for masking, mask flag
        padI = np.fft.fftshift(padI)
        PhaseExtract = np.fft.fft2(padI)
        PhaseExtract = np.fft.fftshift(PhaseExtract)
        AmpPE = np.absolute(PhaseExtract)
        #ArgPE = np.multiply(np.angle(PhaseExtract),180./np.pi) # degrees
        ArgPE = np.angle(PhaseExtract) # radians

        print("Size of array being FFTed:")
        print(np.shape(PhaseExtract))

        if mask:
            # mask out low-power regions
            AmpPE_masked = ma.masked_where(AmpPE < mask_thresh, AmpPE, copy=False)
            ArgPE_masked = ma.masked_where(AmpPE < mask_thresh, ArgPE, copy=False)
            return AmpPE_masked, ArgPE_masked

        else:
            return AmpPE, ArgPE
```
