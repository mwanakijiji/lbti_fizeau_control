```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from astropy.io import fits
import asciitable
import sys
import glob, os
```

```python
# read in a file and extract a meta-data field from the FITS header

dirTreeStem = ('./')
fileNameStem = ('lm_190224_')

os.chdir(dirTreeStem)

counter = 0

for f in glob.glob("*.fits"): # loop over filenames
    counter += 1 # count loops
    
    counterStart = 60
    counterStop = 11500
    if (np.logical_or(counter<counterStart,counter>counterStop)):
        continue
    
    if (counter%50 == 0):
        print('Reading in header info from frame '+str(f)+'...')
    
    # nice continuous printing of status, but can get stuck 
    #print('\r', 'Reading in header info from frame '+str(f)+'...', end='')
    #sys.stdout.flush()
    
    image, header = fits.getdata(dirTreeStem+f,0,header=True)
    
    
    # initialize dictionary with empty lists for each key, if this is first loop iteration
    if (counter==counterStart):
        data_ascii = {}
        for k in header:
            data_ascii[k] = [] 
        data_ascii['NODPOS'] = [] # manually add in header keys that don't appear in all frames
        data_ascii['DATAFLAG'] = []
        data_ascii['BADROWS'] = []
        data_ascii['FRAMENUM'] = []
    

    # append data from frames   
    for l in data_ascii:
        if l in header:
            data_ascii[l].append(header[l])
        else:
            if (l!='FRAMENUM'):
                data_ascii[l].append(np.nan) # if a flag doesn't appear in the header
            else:
                split_underscore = f.split("_") # to get frame number
                #print(split_underscore)
                split_final = split_underscore[2].split(".")
                data_ascii[l].append(split_final[0])
                
    #print(np.shape(data_ascii))
```

```python
# convert to pandas dataframe

data_ascii_df = pd.DataFrame.from_dict(data_ascii, orient='columns')
```

```python
# write to file

data_ascii_df.to_csv('test_0004300.csv')
```

```python
# read in a ready-made csv

stem2 = ("/home/../../media/unasemaje/Seagate Expansion Drive/lbti_data_reduction/"
                "180524_fizeau_ac_her/meta_data/")
df0 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe004300.csv")
df1 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe010000.csv")
df2 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe020000.csv")
df3 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe030000.csv")
df4 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe040000.csv")
df5 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe050000.csv")
df6 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe060000.csv")
df7 = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe070000.csv")
```

```python
# the dataframe, post-concatenation
df_full = pd.read_csv(stem2 + "metadata_180524_acher_uptoframe070000.csv")
```

```python
# concatenate the dataframes

concatenate_all = pd.concat([df0, df1, df2, df3, df4, df5, df6, df7], axis=0, join='outer', join_axes=None, ignore_index=True,
          keys=None, levels=None, names=None, verify_integrity=True,
          copy=True)
meta_data_uniq = concatenate_all.drop_duplicates(subset="FRAMENUM") # remove duplicate rows
meta_data_uniq2 = meta_data_uniq[meta_data_uniq["FRAMENUM"] > 82] # remove unrelated data from earlier in the night
```

```python
# read in nod positions

nod_pos = pd.read_csv("ac_her_psf_coords.csv").drop_duplicates(subset="frame")
```

```python
# rename frame number column for merging with larger meta-data table

nod_pos = nod_pos.rename(index=str, columns={"frame": "FRAMENUM"})
nod_pos_uniq = nod_pos.drop_duplicates(subset="FRAMENUM") # remove duplicate rows
nod_pos_uniq2 = nod_pos_uniq[nod_pos_uniq["FRAMENUM"] > 81] # remove unrelated data from earlier in the night
```

```python
# do the merge

test3 = pd.merge(meta_data_uniq2, nod_pos_uniq2, how="outer", on=["FRAMENUM"])
```

```python
# write to file

meta_data_uniq2.to_csv("metadata_180524_acher_fizeau_frames_frames_geq_83.csv")
```
