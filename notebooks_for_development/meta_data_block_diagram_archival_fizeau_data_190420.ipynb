```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy
from astropy.io import fits
import pandas as pd
from datetime import datetime
#%matplotlib inline
```

```python
# read in subsidiary FITS header data files, then combine them
full_table1 = pd.read_csv('./data/00_meta_data_190420_hd141569.csv')
full_table2 = pd.read_csv('./data/01_meta_data_190420_hd141569.csv')
full_table3 = pd.read_csv('./data/02_meta_data_190420_hd141569.csv')
full_table4 = pd.read_csv('./data/03_meta_data_190420_hd141569.csv')
full_table5 = pd.read_csv('./data/04_meta_data_190420_hd141569.csv')

tables_list = [full_table1, full_table2, full_table3, full_table4, full_table5]

full_table = pd.concat(tables_list)
```

```python
# sort table by frame number (a few are out of order)

full_table.sort_values(["FRAMENUM"], inplace=True)
full_table = full_table.reset_index(drop=True) # reset indices and don't keep the old ones
```
