```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
abcissa = np.linspace(0,10*np.pi,1000)
```

```python
ordinate_small = np.sin(abcissa*10)
ordinate_large = np.sin(abcissa)
```

```python
fig, ax = plt.subplots(figsize=(20,1), frameon=False)
#fig, ax = plt.figure(figsize=(20,0.5), frameon=False)
#ax = fig.add_axes([0, 0, 1, 1])
ax.plot(abcissa, ordinate_small, color="k", linewidth = 4)
#ax.plot(abcissa, ordinate_large, color="k", linewidth = 4)
ax.axis('off')
#ax.axis('off')
fig.savefig("junk.pdf", transparent=True)
```
